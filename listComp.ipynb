```python
string = 'la la la something that rhymes with cool'
```

```python
#3.
numSpaces = sum([1 for char in string if char == ' '])
print(numSpaces)
```

```python
#4.
vowels = 'aeiouAEIOU'
withoutVows = ''.join([char for char in string if char not in vowels])
print(withoutVows)
```

```python
#5.
lessThanFive = ' '.join([word for word in string.split() if len(word) < 5])
print(lessThanFive)
```

```python
#6.
lenWords = {word: len(word) for word in string.split()}
print(lenWords)
```

```python
#7.
divByAny = [num for num in range(1,1001) if any(num % div == 0 for div in range(2,10))]
print(divByAny)
```

```python
#8.
numAndHighest = {num: max([div for div in range(2, 10) if num % div == 0], default=None) for num in range(1, 1001)}
print(numAndHighest)
```
